```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loc1 = pd.read_csv('Loc1_session1_group3_2024-04-12_16-01-55.csv')
loc2 = pd.read_csv('Loc2_session1_group3_2024-04-12_16-11-38.csv')
loc3 = pd.read_csv('Loc3_session1_group3_2024-04-12_16-18-58.csv')
```

## Part A
Note that when we were taking measurements, we only collected radiation data (the counts); no temperature or AQD was collected for some reason, so I will only include plots of the radiation data. Also, our position data showed the same position for all locations, so we can only graph time vs radiation data for this part.

```python
# Time vs. Radiation Data - Location 1
fig = plt.figure()
ax1 = plt.axes()
ax1.plot(loc1['Time'], loc1['counts'])

ax1.set_xlabel('Time (Epoch)')
ax1.set_ylabel('Counts')
```

```python
# Time vs. Radiation Data - Location 2
fig = plt.figure()
ax2 = plt.axes()
ax2.plot(loc2['Time'], loc2['counts'])

ax2.set_xlabel('Time (Epoch)')
ax2.set_ylabel('Counts')
```

```python
# Time vs. Radiation Data - Location 3
fig = plt.figure()
ax3 = plt.axes()
ax3.plot(loc3['Time'], loc3['counts'])

ax3.set_xlabel('Time (Epoch)')
ax3.set_ylabel('Counts')
```

## Part B

```python
# Histogram of Radiation Data - Location 1
fig = plt.figure()
ax4 = plt.axes()
ax4.hist(loc1['counts'], label = 'Location 1', bins = 20, color = 'b')
avgCR1 = np.mean(loc1['counts'])
stdCR1 = np.std(loc1['counts'])
print("The average count value is", avgCR1, "with a standard deviation of", stdCR1)
ax4.set_ylabel('Frequency')


# Histogram of Radiation Data - Location 2
fig = plt.figure()
ax5 = plt.axes()
ax5.hist(loc2['counts'], label = 'Location 2', bins = 20, color = 'b')
avgCR2 = np.mean(loc2['counts'])
stdCR2 = np.std(loc2['counts'])
print("The average count value is", avgCR2, "with a standard deviation of", stdCR2)
ax5.set_ylabel('Frequency')


# Histogram of Radiation Data - Location 3
fig = plt.figure()
ax6 = plt.axes()
ax6.hist(loc3['counts'], label = 'Location 3', bins = 20, color = 'b')
ax6.set_ylabel('Frequency')

avgCR3 = np.mean(loc3['counts'])
stdCR3 = np.std(loc3['counts'])

print("The average count value is", avgCR3, "with a standard deviation of", stdCR3)
```

Each of these data types follow the general shape of a normal distribution / bell curve, although some less than others. This could be because in some locations we could have measured more number of counts that are considered outliers. Another reason is that the number of sample measurements in our locations was still too low for a normal distribution to be approached.

## Part C

```python
# avg mean counts per location - take sqrt to get variance
N_loc1 = np.mean(loc1["counts"])
N_loc2 = np.mean(loc2["counts"])
N_loc3 = np.mean(loc3["counts"])


expected_variance1 = np.sqrt(N_loc1)
expected_variance2 = np.sqrt(N_loc2)
expected_variance3 = np.sqrt(N_loc3)

print("The expected variance of our data from location 1 is", expected_variance1)
print("The expected variance of our data from location 2 is", expected_variance2)
print("The expected variance of our data from location 3 is", expected_variance3)

print("")
measured_variance1 = np.std(loc1["counts"])
measured_variance2 = np.std(loc2["counts"])
measured_variance3 = np.std(loc3["counts"])


print("The measured variance of our data from location 1 is", measured_variance1)
print("The measured variance of our data from location 2 is", measured_variance2)
print("The measured variance of our data from location 3 is", measured_variance3)
```

Comparing the two, we see that the measured variance is lower than the expected variance.
