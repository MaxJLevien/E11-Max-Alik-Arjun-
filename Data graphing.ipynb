```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

inside = pd.read_csv('weather_data_inside.csv')
outside = pd.read_csv('weather_quality_outside.csv')[60:361]

fig = plt.figure()
ax1 = plt.axes()

ax1.plot(inside['Time'], inside['PM1.0'], label = 'PM1.0 for inside data', ls = '--', c = 'r')
ax1.plot(outside['Time'], outside['PM1.0'], label = 'PM1.0 for outside data', ls = '-.', c = 'b')

plt.show()
```
